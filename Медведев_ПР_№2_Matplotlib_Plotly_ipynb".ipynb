```python
import numpy as np
import matplotlib.pyplot as plt
```

1. Подготовьте текстовый файл с двумя столбцами данных, разделёнными пробелами или запятыми. Пусть первый столбец будет временем, а второй - измеренными значениями.
2. Используя `numpy`, загрузите данные из файла с помощью функции `np.loadtxt` или `np.genfromtxt`.
3. Постройте график зависимости измеренных значений от времени.
4. Добавьте заголовок, подписи осей и легенду.

Для примера сгенерируем случайные данные и запишем их в файл data.txt:

```python
import numpy as np
# Создаем массив времени от 0 до 10 секунд с шагом 0.1
time = np.arange(0, 10, 0.1)
# Создаем массив измеренных значений, которые являются суммой синуса и шума
values = np.sin(time) + np.random.normal(0, 0.1, len(time))
# Соединяем два массива в одну матрицу
data = np.column_stack((time, values))
# Записываем матрицу в файл data.txt, разделяя столбцы пробелами
np.savetxt('data.txt', data, delimiter=' ')
```

```python
ds = np.loadtxt('data.txt', delimiter=' ')
fig = plt.figure()
plt.plot(ds[:,:1],ds[:,1:2])
plt.title('График изменения значений от времени')
plt.xlabel('Время')
plt.ylabel('Значение')
```

1. Создайте двумерный массив `x`, используя `numpy.meshgrid`, где значения по обеим осям находятся в диапазоне от -5 до 5.
2. Вычислите $z = \sin(\sqrt{x^2 + y^2})$, где $x$ и $y$ - координаты, полученные с помощью `meshgrid`.
3. Используйте функцию `plt.contourf()` для создания цветного контурного графика функции $z$.
4. Выберите и примените цветовую карту к вашему графику `plt.set_cmap()`.
5. Добавьте цветовую шкалу для интерпретации значений функции `plt.colorbar()`.

```python
x,y = np.meshgrid(np.linspace(-5,5,100),np.linspace(-5,5,100))
def f(x,y):
  return np.sin(np.sqrt(x**2+y**2))
z = f(x,y)
fig = plt.figure()
plt.contourf(z)
plt.set_cmap('rainbow')
plt.colorbar()
```
